```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import re

%matplotlib inline
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# Import Scikit-learn models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

```python
dataset_1_URLs_df = pd.read_csv("/content/Malicious URLs.csv")

dataset_1_URLs_df
```

```python
test_porcentaje = .2

train_df, test_df = train_test_split(dataset_1_URLs_df, test_size=test_porcentaje, random_state=42)

labels = train_df['Class']
test_labels = test_df['Class']

print("\n### Split Complete ###\n")

print("Cantidad de Datos de Entrenamiento")
print(len(train_df))
print("")
print("Cantidad de Datos de Prueba")
print(len(test_df))
```

```python
count_train_classes = pd.value_counts(train_df['Class'])
count_train_classes.plot(kind='bar', fontsize=16)
plt.title("Quantidade de dados de treinamento", fontsize=24)
plt.xticks(rotation='horizontal')
plt.xlabel("Classe", fontsize=15)
plt.ylabel("Quantidade de Dados por Classe", fontsize=15)

plt.show()

count_test_classes = pd.value_counts(test_df['Class'])
count_test_classes.plot(kind='bar', fontsize=16, colormap='ocean')
plt.title("Quantidade de Dados de Teste", fontsize=24)
plt.xticks(rotation='horizontal')
plt.xlabel("Classe", fontsize=15)
plt.ylabel("Quantidade de Dados por Classe", fontsize=15)

plt.show()
```

```python
def tokenizer(url):
  """Separates feature words from the raw data
  Keyword arguments:
    url ---- The full URL

  :Returns -- The tokenized words; returned as a list
  """

  # Split por slash (/) and dash (-)
  tokens = re.split('[/-]', url)

  for i in tokens:
    # Include the splits extensions and subdomains
    if i.find(".") >= 0:
      dot_split = i.split('.')

      # Remove .com and www. since they're too common
      if "com" in dot_split:
        dot_split.remove("com")
      if "www" in dot_split:
        dot_split.remove("www")

      tokens += dot_split

  return tokens

print("\n########## Tokenizer defined ##########\n")
```

```python
test_url = dataset_1_URLs_df['URLs'][0]
print("Imprimimos la URL de un primer ejemplo:")
print(test_url)

print("\n-------------------- Exemplo de URL, retirado como amostra do banco de dados --------------------\n")
print(test_url)

# Tokenize test URL
print("\n-------------------------- Amostra de URL após passar pelo Tokenizer ----------------------------\n")
tokenized_url = tokenizer(test_url)
print(tokenized_url)
```

# Vectorizacion

```python
print("-------------------- Entrenando el Count Vectorizer -")
cVec = CountVectorizer(tokenizer=tokenizer) #Crea el "count vectorizer" llamado "cVec" utilizando la función "CountVectorizer"
count_X = cVec.fit_transform(train_df['URLs']) #Realiza fit_transform en cVec para entrenar el "vectorizer" con las URLs\ de entrenamiento

print("-------------------- Entrenando el TF-IDF Vectorizer -") #Crea el "TF-IDF vectorizer" llamado "tVec" using la función "TfidfVectorizer"
tVec = TfidfVectorizer(tokenizer=tokenizer)
tfidf_X = tVec.fit_transform(train_df['URLs'])


print("\n##### Se Completó la Vectorización !!! #####\n")


print(tfidf_X)
print(count_X)
```

```python
print("- Count Vectorizer -")
test_count_X = cVec.transform(test_df['URLs'])

print("- TFIDF Vectorizer -")
test_tfidf_X = tVec.transform(test_df['URLs'])

print("\n### Vectorizado de los Datos de Prueba Completo ###\n")
```

#GRAFICADOR Y REPORTAR

```python
def generate_report(cmatrix, score, creport):
  """Genera y Muestra informes gráficos.
  Keyword arguments:
    cmatrix - Confusion matrix generated by the model
    score --- Score generated by the model
    creport - Classification Report generated by the model

  :Returns -- N/A
  """

  # Generate confusion matrix heatmap
  plt.figure(figsize=(5,5))
  sns.heatmap(cmatrix,
              annot=True,
              fmt="d",
              linewidths=.5,
              square = True,
              cmap = 'Blues',
              annot_kws={"size": 16},
              xticklabels=['bad', 'good'],
              yticklabels=['bad', 'good'])

  plt.xticks(rotation='horizontal', fontsize=16)
  plt.yticks(rotation='horizontal', fontsize=16)
  plt.xlabel('Rótulo Atual', size=20);
  plt.ylabel('Rótulo Previsto', size=20);

  title = 'Accuracy Score: {0:.4f}'.format(score)
  plt.title(title, size = 20);

  # Display classification report and confusion matrix
  print(creport)
  plt.show()


print("\n### Report Generator Defined ###\n")
```

# MODELO MNB - TFIDF Vectorizer


```python
# Multinomial Naive Bayesian with TF-IDF

#*********************************************************************************
#*************************** Entrenamos el Modelo ********************************
mnb_tfidf = MultinomialNB()
mnb_tfidf.fit(tfidf_X, labels)


# Test the mode (score, predictions, confusion matrix, classification report)
score_mnb_tfidf = mnb_tfidf.score(test_tfidf_X, test_labels)
predictions_mnb_tfidf = mnb_tfidf.predict(test_tfidf_X)
cmatrix_mnb_tfidf = confusion_matrix(predictions_mnb_tfidf, test_labels)
creport_mnb_tfidf = classification_report(predictions_mnb_tfidf, test_labels)

print("\n### Model Built ###\n")
generate_report(cmatrix_mnb_tfidf, score_mnb_tfidf, creport_mnb_tfidf)

print("La suma de todos los datos dan 84093, ya que es la cantidad de test_df")
```

# MODELO REGRESION LOGISTICA LGS - TFIDF Vectorizer

```python
# Logistic Regression con TF-IDF

#*************************** Entrenamos el Modelo ********************************
lgs_tfidf = LogisticRegression(solver='lbfgs')
lgs_tfidf.fit(tfidf_X, labels)

# Test the mode (score, predictions, confusion matrix, classification report)
score_lgs_tfidf = lgs_tfidf.score(test_tfidf_X, test_labels)
predictions_lgs_tfidf = lgs_tfidf.predict(test_tfidf_X)
cmatrix_lgs_tfidf = confusion_matrix(predictions_lgs_tfidf, test_labels)
creport_lgs_tfidf = classification_report(predictions_lgs_tfidf, test_labels)

print("\n### Model Built ###\n")
generate_report(cmatrix_lgs_tfidf, score_lgs_tfidf, creport_lgs_tfidf)
```
